```python
df.columns = [col.replace('.','_') for col in df.columns]
df.columns
```

```python
df = df[['name', 'primaryCategories','reviews_rating','reviews_text',]]
```

```python
df.columns = ['name','category','rating','text'] # szybka zamiana nazw kolumn
```

```python
df.info()
df.describe()
```

```python
df.to_csv('Customer_Reviews_Amazon_clean.csv')
```
